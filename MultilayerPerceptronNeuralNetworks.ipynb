```python
#Realizamos prediccion
predicciones = classifier.predict(imagenes_Entrenamiento)
predicciones.shape
```

```python
from sklearn.metrics import classification_report,confusion_matrix
matrizdeconfucion=confusion_matrix(etiquetas_entrenamiento,predicciones)
print(matrizdeconfucion)
```

```python
#Se pinta mejor la matriz de confusion
#la matriz, se pasa a Dataframe, para poderla graficar
import pandas as pd
cmPd= pd.DataFrame(matrizdeconfucion)
```

```python
#graficamos la matriz de confusion
import seaborn as sn
import matplotlib.pyplot as plt
sn.set(font_scale=0.1) # for label size
sn.heatmap(cmPd, annot=True, annot_kws={"size": 10}, fmt='g', center=0, linewidths=1, cbar=False) # font size
plt.show()
```

```python
#Precision del modelo
from sklearn.metrics import accuracy_score
p = accuracy_score(etiquetas_entrenamiento,predicciones)
print('Precision del modelo')
print()
print(p)
```

**classification_report**

Una funcin de la biblioteca scikit-learn que se utiliza para generar un informe detallado de las
mtricas de evaluacin de un modelo de clasificacin.

Precision:

Indica la capacidad del modelo para predecir correctamente las instancias de una clase especfica.

Recall o Sensibilidad:

Indica la capacidad del modelo para identificar correctamente todas las instancias de una clase
especfica.

F1*Score:

Proporciona una medida general del rendimiento del modelo, considerando tanto la precisin como
el recall.

Support.

Support: Indica el nmero de muestras en cada clase verdadera. Es importante tener en cuenta que,
mientras ms equilibrado sea el conjunto de datos, mayor ser el impacto de esta mtrica en el
clculo del promedio ponderado de las mtricas de precisin, Recall y F1-score.

Adems, este reporte, incluye el promedio de todas las mtricas para todas las clases, que
proporciona una visin general del rendimiento del modelo en todo el conjunto de datos.
