**classification_report**

Una función de la biblioteca scikit-learn que se utiliza para generar un informe detallado de las
métricas de evaluación de un modelo de clasificación.

Precision:

Indica la capacidad del modelo para predecir correctamente las instancias de una clase específica.

Recall o Sensibilidad:

Indica la capacidad del modelo para identificar correctamente todas las instancias de una clase
específica.

F1*Score:

Proporciona una medida general del rendimiento del modelo, considerando tanto la precisión como
el recall.

Support.

Support: Indica el número de muestras en cada clase verdadera. Es importante tener en cuenta que,
mientras más equilibrado sea el conjunto de datos, mayor será el impacto de esta métrica en el
cálculo del promedio ponderado de las métricas de precisión, Recall y F1-score.

Además, este reporte, incluye el promedio de todas las métricas para todas las clases, que
proporciona una visión general del rendimiento del modelo en todo el conjunto de datos.
